```python
# create a column age_group

labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)

# This code is used to create a new categorical column called 'age_group' based on the existing 'age' column in the dataframe. 
# First, a list of labels — 'Young Adult', 'Adult', 'Middle-aged', and 'Senior' — is defined to represent four meaningful age categories. 
# These labels will later be assigned to different segments of the age data. Then, the pandas function pd.qcut() is applied to the 'age' column
# with the parameter q=4, which divides the age values into four equal-sized groups based on their statistical distribution (quartiles), 
# ensuring that each group contains approximately 25% of the total observations. Instead of returning numerical age ranges, the function assigns 
# the predefined category labels to each group, making the results more interpretable and useful for analysis. The final categorized values are 
# stored in a newly created column named 'age_group' in the dataframe for further exploratory and business analysis.
```

```python
#Create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)

#This code converts the categorical purchase frequency information into numerical values representing days. 
# First, a dictionary named frequency_mapping is created, where each textual purchase frequency category 
# (such as 'Weekly', 'Monthly', 'Quarterly', etc.) is assigned a corresponding number of days 
# (for example, 'Weekly' is mapped to 7 days and 'Monthly' to 30 days). This dictionary acts as a reference lookup table for conversion. 
# Then, the .map() function is applied to the 'frequency_of_purchases' column, which replaces each category with its mapped numerical 
# value from the dictionary. The converted values are stored in a new column called 'purchase_frequency_days', making the data suitable 
# for quantitative analysis, calculations, and further modeling.
    
```

```python
from sqlalchemy import create_engine

#sqlalchemy = a library that helps Python talk to databases
#create_engine = a tool that creates the connection to PostgreSQL


# Step 1: Connect to PostgreSQL
# Replace placeholders with your actual details

username = "postgres"                          #default user
password = "Admin123"                          # the password you set during installation  
host = "localhost"                             # if running locally
port = "5432"                                  # default PostgreSQL port
database = "customer_behavior"                 # the database you created in pgAdmin

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# This line creates a connection engine that acts as a bridge between Python and PostgreSQL. 
# The connection string includes the database type (postgresql+psycopg2), username, password,
# host, port, and database name. This engine will be used later to send data from the Pandas DataFrame to PostgreSQL.


# Step 2: Load DataFrame into PostgreSQL
table_name = "customer"                                            # choose any table name or You are choosing a table name in your database.
df.to_sql(table_name, engine, if_exists="replace", index=False)

# This line uploads the Pandas DataFrame df into the PostgreSQL database as a table named customer. 
# The parameter if_exists="replace" means that if the table already exists, it will be replaced with the new data. 
# The parameter index=False ensures that the DataFrame index is not saved as a separate column in the database.

# df = your entire dataset (DataFrame)
# to_sql() = function that sends data to SQL database
# engine = your database connection
# if_exists="replace" = overwrite old table
# index=False = don’t save row numbers as a column


print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")

#This is just a success message so you know everything worked.
```
